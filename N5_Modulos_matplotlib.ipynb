## 4. Cargar Modulos. Matplotlib

Python tiene definidos modulos que son conjuntos de funciones. Las funciones se acceden: modulo.funcion() 


```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
# puede leer archivos
A=np.loadtxt('ejemplo.txt')
print('dimensiones de A',np.shape(A))
```

```python
x=A[:,0]
y=A[:,1]
```

```python
import matplotlib.pyplot as pl
pl.plot(x,y)
pl.show()
```

```python
pl.plot(x,y)
pl.xlabel('eje x')
pl.ylabel('eje y')
pl.show()
```

```python
pl.plot(x,y)
pl.xlabel('eje x')
pl.ylabel('eje y')
pl.show()
```

```python
pl.plot(x,y,'*')
pl.xlabel('eje x')
pl.ylabel('eje y')
```

```python
pl.plot(x,y,'*',color='r')
pl.plot(x,2*y,'*',color='b')
pl.xlabel('eje x')
pl.ylabel('eje y')
```

```python
pl.plot(x,y,'*',color='r',label='datos 1')
pl.plot(x,2*y,'^',color='b',label='datos 2')
pl.xlabel('eje x')
pl.ylabel('eje y')
pl.legend()
```
