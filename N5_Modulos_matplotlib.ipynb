## 4. Cargar Modulos. Matplotlib

Python tiene definidos modulos que son conjuntos de funciones. Las funciones se acceden: modulo.funcion() 


```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
# puede leer archivos
A=np.loadtxt('ejemplo.txt')
print('dimensiones de A',np.shape(A))
```

```python
x=A[:,0]
y=A[:,1]
```

```python
import matplotlib.pyplot as pl
pl.plot(x,y)
pl.show()
```

```python
pl.plot(x,y)
pl.xlabel('eje x')
pl.ylabel('eje y')
pl.show()
```

```python
pl.plot(x,y)
pl.xlabel('eje x')
pl.ylabel('eje y')
pl.show()
```

```python
pl.plot(x,y,'*')
pl.xlabel('eje x')
pl.ylabel('eje y')
```

```python
pl.plot(x,y,'*',color='r')
pl.plot(x,2*y,'*',color='b')
pl.xlabel('eje x')
pl.ylabel('eje y')
```

```python
pl.plot(x,y,'*',color='r',label='datos 1')
pl.plot(x,2*y,'^',color='b',label='datos 2')
pl.xlabel('eje x')
pl.ylabel('eje y')
pl.legend()
```

```python
# y también se pueden cambiar los límiter
pl.plot(x,y,'*-',color='r',label='datos 1')
pl.plot(x,2*y,'^-',color='b',label='datos 2')
pl.xlabel('eje x')
pl.ylabel('eje y')
pl.legend()
pl.xlim(2,8)
pl.ylim(5,30)
```

```python
# y también lo podríamos poner en graficos distintos
pl.subplot(1,2,1)
pl.plot(x,y,'*',color='r',label='datos 1')
pl.subplot(1,2,2)
pl.plot(x,2*y,'^',color='b',label='datos 2')
pl.xlabel('eje x')
pl.ylabel('eje y')
pl.legend()
```

```python
a=np.random.normal(size=100)
pl.hist(a)
```

```python
# y se pueden hacer cosas más complicadas
fig = pl.figure()
ax = fig.add_axes([0.0, 0.0, .6, .6], polar=True)
t = np.linspace(0, 2 * np.pi, 100)
ax.plot(t, t, color='blue', lw=3);
```

```python
fig = pl.figure()
ax1 = pl.subplot2grid((3,3), (0,0), colspan=3)
#pl.plot(x,x)
ax2 = pl.subplot2grid((3,3), (1,0), colspan=2)
ax3 = pl.subplot2grid((3,3), (1,2), rowspan=2)
ax4 = pl.subplot2grid((3,3), (2,0))
ax5 = pl.subplot2grid((3,3), (2,1))
fig.tight_layout()
```
